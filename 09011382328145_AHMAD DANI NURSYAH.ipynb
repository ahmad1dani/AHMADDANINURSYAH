Penjelasan Skewness:
Skewness adalah ukuran statistik yang menggambarkan simetri distribusi data di sekitar rata-rata. Ada beberapa jenis skewness:

Skewness Positif (Right Skewed):

Nilai: Skewness > 0
Makna: Distribusi data condong ke arah kiri (lebih banyak nilai kecil) dan memiliki ekor yang lebih panjang di sisi kanan. Ini berarti bahwa sebagian besar data terletak di sisi kiri dan ada beberapa nilai ekstrem di sisi kanan.
Skewness Negatif (Left Skewed):

Nilai: Skewness < 0
Makna: Distribusi data condong ke arah kanan (lebih banyak nilai besar) dan memiliki ekor yang lebih panjang di sisi kiri. Ini berarti bahwa sebagian besar data terletak di sisi kanan dan ada beberapa nilai ekstrem di sisi kiri.
Skewness Nol (Symmetric):

Nilai: Skewness mendekati 0
Makna: Distribusi data simetris, yang berarti data terdistribusi merata di sekitar rata-rata, dan tidak ada kemiringan yang signifikan ke kiri atau kanan.

max_heart_rate: Nilai skewness -0.678 menunjukkan distribusi detak jantung maksimum yang condong ke kiri dengan ekor panjang di sisi kiri, menunjukkan bahwa ada beberapa nilai detak jantung yang sangat rendah.

Penjelasan Kurtosis:
Kurtosis adalah ukuran statistik yang menggambarkan bentuk distribusi data, khususnya ekor distribusi dan puncaknya. Ada beberapa jenis kurtosis, tetapi yang paling umum adalah:

Kurtosis Positif (Leptokurtik): Nilai kurtosis lebih besar dari 3 (kurtosis normal). Distribusi ini memiliki puncak yang tajam dan ekor yang lebih berat, menunjukkan lebih banyak data ekstrem atau outlier.

Penjelasan Umum Hasil Histogram:
Histogram adalah alat visualisasi yang menunjukkan distribusi frekuensi dari nilai-nilai numerik. Berikut adalah beberapa elemen penting yang dapat dianalisis dari histogram:

Distribusi Data:

Histogram menunjukkan bagaimana data terdistribusi. Kamu dapat melihat sebaran data di sepanjang sumbu x (nilai variabel) dan frekuensinya di sumbu y.
Kepadatan Data:

Jumlah bar (bin) dalam histogram menunjukkan frekuensi atau jumlah data dalam rentang nilai tertentu. Bar yang lebih tinggi menunjukkan bahwa ada lebih banyak data dalam rentang tersebut.
Skewness (Kemiringan):

Jika histogram condong ke kiri (skewed left) atau ke kanan (skewed right), ini dapat menunjukkan bahwa data memiliki distribusi yang tidak simetris.
Puncak (Mode):

Lokasi puncak atau bar tertinggi dalam histogram menunjukkan nilai-nilai yang paling sering muncul dalam data. Ini memberi informasi tentang nilai-nilai dominan dalam dataset.
Outliers:

Histogram juga dapat menunjukkan adanya nilai-nilai yang sangat tinggi atau rendah dibandingkan dengan mayoritas data. Bar yang jauh dari distribusi utama bisa menunjukkan adanya outliers.
Rentang Nilai:

Dengan melihat histogram, kamu dapat melihat rentang nilai variabel dan bagaimana data tersebar di sepanjang rentang tersebut.
Contoh Bacaan Hasil Histogram:
Misalkan kamu memiliki histogram untuk kolom age dalam dataset penyakit jantung:

Distribusi Umum: Jika histogram menunjukkan bahwa sebagian besar data terpusat pada usia tertentu (misalnya, antara 50 hingga 60 tahun), ini menunjukkan bahwa usia mayoritas pasien cenderung berada dalam rentang tersebut.

Skewness: Jika histogram menunjukkan bahwa lebih banyak data berada di usia yang lebih muda dan sangat sedikit di usia yang lebih tua, ini menunjukkan distribusi data yang skewed ke kiri.

Puncak: Jika histogram memiliki puncak di usia 55 tahun, maka usia 55 tahun adalah usia yang paling umum atau sering terjadi dalam dataset.

Outliers: Jika ada beberapa bar yang jauh di luar rentang utama, ini menunjukkan adanya outliers di usia yang sangat muda atau sangat tua.

Rentang Nilai: Jika histogram menunjukkan rentang usia dari 20 hingga 80 tahun, maka data mencakup usia dalam rentang tersebut.

Jika kamu memiliki hasil histogram yang spesifik dan ingin menjelaskan atau menganalisis hasil tersebut, pastikan untuk memberikan rincian atau gambar histogram tersebut sehingga analisis yang lebih terperinci dapat dilakukan.








```python

# Pilih variabel numerik
numerical_data = data.select_dtypes(include=['number'])

# Hitung matriks korelasi
correlation_matrix = numerical_data.corr()

# Menghilangkan korelasi dengan dirinya sendiri (nilai diagonal) dan bagian atas matriks
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool), k=1)
masked_corr = correlation_matrix.where(~mask)

# Temukan pasangan variabel dengan korelasi tertinggi
max_corr = masked_corr.stack().idxmax()
max_corr_value = masked_corr.stack().max()

# Ambil nama variabel yang berkorelasi tinggi
var1, var2 = max_corr

# Buat scatter plot untuk pasangan variabel dengan korelasi tinggi
plt.figure(figsize=(10, 6))
sns.scatterplot(data=numerical_data, x=var1, y=var2)
plt.title(f'Scatter Plot antara {var1} dan {var2} (Korelasi: {max_corr_value:.2f}) - Data Penyakit Jantung UCI')
plt.ylabel(var2)
plt.show()
```
